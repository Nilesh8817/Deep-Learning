```python
df.drop([2207,2314,4791,5471,10167,11182,11963,13601,15951,24394,24520,26123,26243,27112,27429,27503,27739,49556,49557,51506],inplace=True)
```

```python
df_final = pd.concat([df_cat,df_num],axis=1)
df_final.head()
```

```python
df_final.isnull().sum()
```

```python
from scipy.stats import skew

skew =df_final["price"].skew()
skew
```

```python
df_final["price"] = np.log(df_final["price"])
```

```python
sk=df_final["price"].skew()
sk
```

```python
X = df_final.drop("price",axis=1)
y = df_final["price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train_ss = ss.fit_transform(X_train)
X_test_ss = ss.transform(X_test)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
       tf.keras.layers.Dense(1, input_shape=(9,))                             
])
```

```python
model.summary()
```

```python
model.compile(loss="mse", optimizer="sgd")
```

```python
history = model.fit(X_train_ss,y_train,epochs=20, batch_size=30)
```

```python
plt.plot(history.history["loss"])
```

```python
y_pred = model.predict(X_test_ss)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test,y_pred)
r2= r2_score(y_test,y_pred)

print("mse:",mse)
print("rmse:",rmse)
print("mae:",mae)
print("r2:",r2)
```

```python
# so we get train MSE as 0.05 and test MSE as 0.04 which is low and good.
```
